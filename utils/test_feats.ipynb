```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from mpl_toolkits.mplot3d import axes3d

from IPython.display import display, Markdown
from ipywidgets import widgets, interactive
```

```python
%matplotlib notebook
```

```python
data, clusters = make_blobs(1000, 3, random_state=100)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=clusters)
```

```python
options = ['test {}'.format(i) for i in range(1,6)]

def widget_test(selection):
    display(Markdown('#### {}'.format(list(selection))))

m = interactive(widget_test, selection=widgets.SelectMultiple(options=options, 
                                                         rows=5,
                                                         value=[options[0]], 
                                                         description='Genre:'))

display(Markdown('#### Chose test'))
display(m)
```
